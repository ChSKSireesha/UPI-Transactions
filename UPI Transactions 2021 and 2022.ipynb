```python
# Using the user defined function convert the ‘Value (Cr) by Costumers’ column into a categorical column as follow-
#a) If the amount is less than 5 Millions then recode as TypeA.
#b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
#c) If the amount is greater than 7 Millions then recode as TypeC.
```

```python
def value_by_customers(value):
    if value<5:
        return "TypeA"
    elif (value>5) & (value<7):
        return "TypeB"
    elif (value>7):
        return "TypeC"
```

```python
main_data_frame["Value (Cr) by Costumers"].apply(value_by_customers)
```

```python
#Group the data by UPI banks and calculate the total transaction volume and value for each bank.

main_data_frame.groupby("UPI Banks")[["Volume (Mn) By Costumers","Value (Cr) by Costumers"]].sum()
```

```python
#Create a pivot table to analyze the transaction volume and value based on  month.
main_data_frame.pivot_table(values = ["Volume (Mn) By Costumers","Value (Cr) by Costumers"], index = ["Month_x"])
```

```python
# Identify months where a particular bank had a transaction volume greater than a certain threshold.

per = (df_2021["Value (Cr) by Costumers"]/sum(df_2021["Value (Cr) by Costumers"]))*100
threshold = 0.002
df_2021.loc[per > threshold, ["UPI Banks", "Month", "Value (Cr) by Costumers"]]
```

```python
#What is the average transaction value for customers in a specific bank in a given month?

df_2021.groupby(["UPI Banks","Month"])[["Value (Cr) by Costumers"]].mean()
```

```python
#Which UPI banks had the highest transaction volume and value in each month?

df_2021.groupby(["UPI Banks","Month"])[["Value (Cr) by Costumers","Volume (Mn) By Costumers"]].max()
```
